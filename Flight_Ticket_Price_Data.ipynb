```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import requests
import time
from selenium.common.exceptions import NoSuchElementException
```

Imported necessary libraries

```python
driver = webdriver.Chrome(r'C:\Users\sawan\Desktop\chromedriver.exe')
```

```python
url = 'https://www.easemytrip.com/'
driver.get(url)
```

Using easemytrip.com to scrap the data

```python
url_urls=[]
```

```python
clickss=driver.find_elements(By.XPATH,'//a[@class="srch_btn"]')
for i in clickss:
    url_urls.append(i.get_attribute("href"))
```

Collected URLs to scrap the data

```python
url_urls
```

```python
heading9 = []
ae = range(11,21)

for url in url_urls:
    driver.get(url)
    time.sleep(5) 
        
    for ai in ae:
        ab = ai
        ac = (str(ab) + '/10/2022')
        ad = str(ac)
        af = "//span[@id="+ '"' +ad+ '"' +"]"
        
        try:        
            Nc0click = driver.find_element(By.XPATH,'/html/body/form/div[9]/div[2]/div/div[4]/div/div/div[1]/div/div[1]/div[2]/div[1]/input[1]')
            Nc0click.click()
        except NoSuchElementException:
            break        
        
        Nc01click = driver.find_element(By.XPATH,af)
        Nc01click.click()

        Nc5lick = driver.find_element(By.XPATH,'/html/body/form/div[9]/div[2]/div/div[4]/div/div/div[1]/div/div[1]/div[3]/button[1]')
        Nc5lick.click()

        time.sleep(5)

        Nclick = driver.find_element(By.XPATH,'/html/body/form/div[9]/div[5]/div/div[1]/div/div/div[13]/div[3]/div[1]/div/p')
        Nclick.click() 

        time.sleep(2) 

        try:
            column9 = driver.find_elements(By.XPATH,'//div[@class="col-md-12 col-sm-12 main-bo-lis pad-top-bot ng-scope"]')    
            for z in column9:
                heading9.append(z.text) 
        except NoSuchElementException:
            heading9.append('-') 

        time.sleep(1)

        Nclick = driver.find_element(By.XPATH,'/html/body/form/div[9]/div[5]/div/div[1]/div/div/div[13]/div[3]/div[1]/div/p')
        Nclick.click()

        Fclick = driver.find_element(By.XPATH,'/html/body/form/div[9]/div[5]/div/div[1]/div/div/div[13]/div[3]/div[2]/div/p')
        Fclick.click()

        time.sleep(2)

        try:
            column9 = driver.find_elements(By.XPATH,'//div[@class="col-md-12 col-sm-12 main-bo-lis pad-top-bot ng-scope"]')    
            for z in column9:
                heading9.append(z.text) 
        except NoSuchElementException:
            heading9.append('-') 

        time.sleep(1)

        Fclick = driver.find_element(By.XPATH,'/html/body/form/div[9]/div[5]/div/div[1]/div/div/div[13]/div[3]/div[2]/div/p')
        Fclick.click()

        Sclick = driver.find_element(By.XPATH,'/html/body/form/div[9]/div[5]/div/div[1]/div/div/div[13]/div[3]/div[3]/div/p')
        Sclick.click()

        time.sleep(2)

        try:
            column9 = driver.find_elements(By.XPATH,'//div[@class="col-md-12 col-sm-12 main-bo-lis pad-top-bot ng-scope"]')    
            for z in column9:
                heading9.append(z.text) 
        except NoSuchElementException:
            heading9.append('-') 

```
